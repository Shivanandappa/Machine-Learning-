```python
def mae(predictions,actual):
    absolute_errors = [abs(x-y)for x,y in zip(predictions,actual)]
    sum_of_absolute_errors = sum(absolute_errors)
    return sum_of_absolute_errors / len(absolute_errors)
```

```python
mae(
    [-3,0,3],
    [0,0,0]
)
```

```python
#another way
def mae(predictions,actual):
    absolute_errors = []
    for i in range(0,len(predictions)):
        x = predictions[i]
        y = actual[i]
        error = abs(x-y)
        absolute_errors.append(error)
    sum_of_absolute_errors = sum(absolute_errors)
    return sum_of_absolute_errors / len(absolute_errors)
```

```python
mae(
    [-3,0,3],
    [0,0,0]
)
```

##Root mean square error

```python
def rmse(predictions,actual):
    squared_errors = [(x-y) ** 2 for x,y in zip(predictions,actual)]
    sum_of_squared_errors = sum(squared_errors)
    return (sum_of_squared_errors/len(squared_errors))**0.5
```

```python
rmse(
    [-3,0,3],
    [0,0,0]   
)
```
